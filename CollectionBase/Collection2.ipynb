# 2. Keypoint extraction using MediaPipe

```python
mp_holistic = mp.solutions.holistic # Holistic model
mp_drawing = mp.solutions.drawing_utils # Drawing utilities

def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # COLOR CONVERSION BGR 2 RGB
    image.flags.writeable = False                  # Image is no longer writeable
    results = model.process(image)                 # Make prediction
    image.flags.writeable = True                   # Image is now writeable 
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # COLOR COVERSION RGB 2 BGR
    return image, results

def draw_landmarks(image, results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION) # Draw face connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS) # Draw pose connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS) # Draw right hand connections

def draw_styled_landmarks(image, results):
    # Draw face connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                              mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                              mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                              )
    # Draw pose connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                              )
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                              )
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                              mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                              mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                              )
cap = cv2.VideoCapture(5)
# Set mediapipe model 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():

        # Read feed
        ret, frame = cap.read()

        # Mirror the frame
        frame = cv2.flip(frame, 1)

        # Make detections
        image, results = mediapipe_detection(frame, holistic)
        print(results)

        # Draw landmarks
        draw_styled_landmarks(image, results)

        # Show to screen
        cv2.imshow('OpenCV Feed', image)

        # Break gracefully#
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        cap.release()
        cv2.destroyAllWindows()
        cv2.namedWindow('OpenCV Feed', cv2.WINDOW_NORMAL)
cv2.resizeWindow('OpenCV Feed', 1280, 720)
```

```python
import cv2
import time
import re
import os
import numpy as np
import threading

# Initialize the variable for user input
user_input = ''

cap = cv2.VideoCapture(5)
cap.set(3, 1920)  # Set frame width to 1920 pixels
cap.set(4, 1080)  # Set frame height to 1080 pixels
cap.set(5, 60)    # Set frames per second to 60

def extract_and_save(v_image, v_results, npy_path_with_face, npy_path_no_face):
    keypoints_with_face = extract_keypoints(v_results, [True, True, True, True])
    keypoints_no_face = extract_keypoints(v_results, [True, False, True, True])
    cv2.imwrite(npy_path_with_face.replace('.npy', '.jpg'), v_image)
    np.save(npy_path_with_face, keypoints_with_face)
    np.save(npy_path_no_face, keypoints_no_face)

def capture_frames(p_action, p_sequence, holistic, sequence_length):
    for frame_num in range(sequence_length):
        v_ret, v_frame = cap.read()
        v_frame = cv2.flip(v_frame, 1)
        v_image, v_results = mediapipe_detection(v_frame, holistic)
        draw_styled_landmarks(v_image, v_results)

        # Display starting text on the first frame
        if frame_num == 0:
            cv2.putText(v_image, 'STARTING COLLECTION', (120, 200),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
        cv2.putText(v_image, 'Collecting frames for {} Sequence Number {}'.format(p_action, p_sequence), (15, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

        # Show the current frame on the screen
        cv2.imshow('OpenCV Feed', v_image)
        cv2.waitKey(10)

        npy_path_with_face = os.path.join(DATA_PATH, p_action, str(p_sequence), str(frame_num) + '.npy')
        npy_path_no_face = os.path.join(DATA_PATH_NO_FACE, p_action, str(p_sequence), str(frame_num) + '.npy')

        # Handle extraction and saving in a separate thread
        threading.Thread(target=extract_and_save, args=(v_image, v_results, npy_path_with_face, npy_path_no_face)).start()

        # Break gracefully if 'q' is pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Set mediapipe model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    # Loop through actions
    for action in actions:
        # Loop through sequences aka videos
        for sequence in range(start_folder, start_folder + no_sequences):
            while True:
                ret, frame = cap.read()
                frame = cv2.flip(frame, 1)
                # Make detections
                image, results = mediapipe_detection(frame, holistic)
                cv2.putText(image, 'Next Action: {}, Sequence: {}'.format(action, sequence), (10, 50),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                # Draw landmarks
                draw_styled_landmarks(image, results)

                cv2.imshow('OpenCV Feed', image)
                key = cv2.waitKey(10) & 0xFF
                if key == ord('g'):
                    capture_frames(action, sequence, holistic, sequence_length)
                    break
                elif key == ord('h'):
                    for i in range(3, 0, -1):
                        ret, frame = cap.read()
                        frame = cv2.flip(frame, 1)

                        # Make detections
                        image, results = mediapipe_detection(frame, holistic)

                        # Draw landmarks
                        draw_styled_landmarks(image, results)

                        cv2.putText(image, 'Starting in {}'.format(i), (120, 200),
                                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
                        cv2.imshow('OpenCV Feed', image)
                        cv2.waitKey(1000)
                    capture_frames(action, sequence, holistic, sequence_length)
                    break
                elif key == ord('c'):
                    retake_pattern = r"([^,\s]+),\b(([0-9]|1[0-9]|2[0-9]))\b"
                    user_input = input("Enter your text: ")
                    print("User input: ", user_input)
                    match = re.search(retake_pattern, user_input)
                    if match:
                        action = match.group(1)
                        sequence = int(match.group(2))
                        for i in range(3, 0, -1):
                            ret, frame = cap.read()
                            frame = cv2.flip(frame, 1)

                            # Make detections
                            image, results = mediapipe_detection(frame, holistic)

                            # Draw landmarks
                            draw_styled_landmarks(image, results)

                            cv2.putText(image, 'Starting in {}'.format(i), (120, 200),
                                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
                            cv2.imshow('OpenCV Feed', image)
                            cv2.waitKey(1000)
                        capture_frames(action, sequence, holistic, sequence_length)
                        break
                    else:
                        print("Invalid input. Please try again.")
                elif key == ord('q'):
                    break_flag = True
                    break

            if 'break_flag' in locals() and break_flag:
                break

    cap.release()
    cv2.destroyAllWindows()

cap.release()
cv2.destroyAllWindows()
```

## Image only collection, post processing

```python
collected_images = []
collected_results = []
collected_paths_with_face = []
collected_paths_no_face = []
```

```python
def extract_and_save(v_image, v_results, npy_path_with_face, npy_path_no_face):
    keypoints_with_face = extract_keypoints(v_results, [True, True, True, True])
    keypoints_no_face = extract_keypoints(v_results, [True, False, True, True])

    # Save keypoints
    np.save(npy_path_with_face, keypoints_with_face)
    np.save(npy_path_no_face, keypoints_no_face)
    con = np.load(npy_path_with_face)

    # Render keypoints on the image
    rendered_image_with_face = v_image.copy()
    rendered_image_no_face = v_image.copy()

    draw_styled_landmarks(rendered_image_with_face, v_results)
    draw_styled_landmarks(rendered_image_no_face, v_results)

    # Save the images with rendered keypoints
    cv2.imwrite(npy_path_with_face.replace('.npy', '_with_face.jpg'), rendered_image_with_face)
    cv2.imwrite(npy_path_no_face.replace('.npy', '_no_face.jpg'), rendered_image_no_face)

def capture_frames(p_action, p_sequence, holistic, sequence_length):
    global collected_images, collected_results, collected_paths_with_face, collected_paths_no_face
    for frame_num in range(sequence_length):
        v_ret, v_frame = cap.read()
        v_frame = cv2.flip(v_frame, 1)
        v_image, v_results = mediapipe_detection(v_frame, holistic)
        

        # Display starting text on the first frame
        if frame_num == 0:
            cv2.putText(v_image, 'STARTING COLLECTION', (120, 200),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
        cv2.putText(v_image, 'Collecting frames for {} Sequence Number {}'.format(p_action, p_sequence), (15, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

        # Show the current frame on the screen
        cv2.imshow('OpenCV Feed', v_image)
        cv2.waitKey(10)

        collected_images.append(v_image)
        collected_results.append(v_results)
        npy_path_with_face = os.path.join(DATA_PATH, p_action, str(p_sequence), str(frame_num) + '.npy')
        npy_path_no_face = os.path.join(DATA_PATH_NO_FACE, p_action, str(p_sequence), str(frame_num) + '.npy')

        # Collect images and paths
        collected_paths_with_face.append(npy_path_with_face)
        collected_paths_no_face.append(npy_path_no_face)

        # Break gracefully if 'q' is pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

def process_collected_images():
    global collected_images, collected_results, collected_paths_with_face, collected_paths_no_face
    threads = []
    for image, results, path_with_face, path_no_face in zip(collected_images, collected_results, collected_paths_with_face, collected_paths_no_face):
        thread = threading.Thread(target=extract_and_save, args=(image, results, path_with_face, path_no_face))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
```

```python
import cv2
import os
import numpy as np
import threading
import re
from scipy.signal import wiener

# Initialize the variable for user input
user_input = ''
def apply_wiener_filter(img, kernel_size=5):
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    filtered_image = wiener(img, (kernel_size, kernel_size))
    filtered_image = np.uint8(filtered_image)
    return filtered_image

def extract_and_save(v_image, npy_path_with_face, npy_path_no_face):
    #Apply wiener to v_image
    v_image = cv2.cvtColor(v_image, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(npy_path_with_face.replace('.npy', '_vanilla_with_face.jpg'), v_image)
    cv2.imwrite(npy_path_no_face.replace('.npy', 'vanilla_no_face.jpg'), v_image)
    v_image = apply_wiener_filter(v_image) # De-blurring the image
    cv2.imwrite(npy_path_with_face.replace('.npy', '_wiener_with_face.jpg'), v_image)
    cv2.imwrite(npy_path_no_face.replace('.npy', '_wiener_no_face.jpg'), v_image)
            
    
    v_results = mediapipe_detection(v_image, holistic)
    keypoints_with_face = extract_keypoints(v_results, [True, True, True, True])
    keypoints_no_face = extract_keypoints(v_results, [True, False, True, True])

    # Save keypoints
    np.save(npy_path_with_face, keypoints_with_face)
    np.save(npy_path_no_face, keypoints_no_face)

    # Render keypoints on the image
    rendered_image_with_face = v_image.copy()
    rendered_image_no_face = v_image.copy()

    draw_styled_landmarks(rendered_image_with_face, v_results)
    draw_styled_landmarks(rendered_image_no_face, v_results)

    # Save the images with rendered keypoints
    cv2.imwrite(npy_path_with_face.replace('.npy', '_with_face_2.jpg'), rendered_image_with_face)
    cv2.imwrite(npy_path_no_face.replace('.npy', '_no_face_2.jpg'), rendered_image_no_face)

def capture_frames(p_action, p_sequence, holistic, sequence_length):
    global collected_images, collected_paths_with_face, collected_paths_no_face
    for frame_num in range(sequence_length):
        v_ret, v_frame = cap.read()
        v_frame = cv2.flip(v_frame, 1)
        
        collected_images.append(v_frame.copy())  # Save original frame without text and mediapipe detection

        # Show the current frame on the screen with overlay
        cv2.putText(v_frame, 'Collecting frames for {} Sequence Number {}'.format(p_action, p_sequence), (15, 12),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
        cv2.imshow('OpenCV Feed', v_frame)
        cv2.waitKey(10)

        npy_path_with_face = os.path.join(DATA_PATH, p_action, str(p_sequence), str(frame_num) + '.npy')
        npy_path_no_face = os.path.join(DATA_PATH_NO_FACE, p_action, str(p_sequence), str(frame_num) + '.npy')
        collected_paths_with_face.append(npy_path_with_face)
        collected_paths_no_face.append(npy_path_no_face)
        # Break gracefully if 'q' is pressed
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

def process_collected_images():
    global collected_images, collected_paths_with_face, collected_paths_no_face
    threads = []
    for image, path_with_face, path_no_face in zip(collected_images, collected_paths_with_face, collected_paths_no_face):
        thread = threading.Thread(target=extract_and_save, args=(image, path_with_face, path_no_face))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()

cap = cv2.VideoCapture(4)
cap.set(3, 1920)  # Set frame width to 1920 pixels
cap.set(4, 1080)  # Set frame height to 1080 pixels
cap.set(5, 60)    # Set frames per second to 60

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    # Loop through actions
    for action in actions:
        # Loop through sequences aka videos
        for sequence in range(start_folder, start_folder + no_sequences):
            while True:
                ret, frame = cap.read()
                frame = cv2.flip(frame, 1)
                # Make detections
            #    image, results = mediapipe_detection(frame, holistic)
                cv2.putText(frame, 'Next Action: {}, Sequence: {}'.format(action, sequence), (10, 50),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

                cv2.imshow('OpenCV Feed', frame)
                key = cv2.waitKey(10) & 0xFF
                if key == ord('g'):
                    capture_frames(action, sequence, holistic, sequence_length)
                    break
                elif key == ord('h'):
                    for i in range(3, 0, -1):
                        ret, frame = cap.read()
                        frame = cv2.flip(frame, 1)

                        # Make detections
                      #  image, results = mediapipe_detection(frame, holistic)

                        cv2.putText(frame, 'Starting in {}'.format(i), (120, 200),
                                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
                        cv2.imshow('OpenCV Feed', frame)
                        cv2.waitKey(1000)
                    capture_frames(action, sequence, holistic, sequence_length)
                    break
                elif key == ord('c'):
                    retake_pattern = r"([^,\s]+),\b(([0-9]|1[0-9]|2[0-9]))\b"
                    user_input = input("Enter your text: ")
                    print("User input: ", user_input)
                    match = re.search(retake_pattern, user_input)
                    if match:
                        action = match.group(1)
                        sequence = int(match.group(2))
                        for i in range(3, 0, -1):
                            ret, frame = cap.read()
                            frame = cv2.flip(frame, 1)

                            # Make detections
                        #    image, results = mediapipe_detection(frame, holistic)

                            cv2.putText(frame, 'Starting in {}'.format(i), (120, 200),
                                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 4, cv2.LINE_AA)
                            cv2.imshow('OpenCV Feed', frame)
                            cv2.waitKey(1000)
                        capture_frames(action, sequence, holistic, sequence_length)
                        break
                    else:
                        print("Invalid input. Please try again.")
                elif key == ord('q'):
                    break_flag = True
                    break

            if 'break_flag' in locals() and break_flag:
                break

    cap.release()
    cv2.destroyAllWindows()
```
